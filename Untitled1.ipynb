```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor as DTR
from sklearn.model_selection import train_test_split as tts
from lightgbm import LGBMRegressor
import lightgbm as lgb
import numpy as np
import datetime
```

```python
df = pd.read_csv("train(1).csv")
df["datetime"] = pd.to_datetime(df["datetime"])
df.head()
```

```python
df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df.head()
```

```python
y = df["y"]
x = df.drop(["y", "datetime"], axis=1)
x.head()
```

```python
dtr = DTR()
dtr.fit(x, y)
imp = dtr.feature_importances_
df_imp = pd.DataFrame(imp)
df_imp.columns = ["imp"]
df_imp.index = x.columns
df_imp.sort_values("imp", ascending=False)
```

```python
models = []
for i in range(400):
    x_train, x_test, y_train, y_test = tts(x, y, test_size=0.3, random_state=i)
    x_train1, x_val1, y_train1, y_val1 = tts(x_train, y_train, test_size=0.2, random_state=0)
    x_train2, x_val2, y_train2, y_val2 = tts(x_train, y_train, test_size=0.2, random_state=1)
    x_train3, x_val3, y_train3, y_val3 = tts(x_train, y_train, test_size=0.2, random_state=2)
    x_train4, x_val4, y_train4, y_val4 = tts(x_train, y_train, test_size=0.2, random_state=3)
    model1 = LGBMRegressor()
    model1.fit(x_train1, y_train1, 
              callbacks=[lgb.early_stopping(stopping_rounds=10, verbose=True)],
              eval_set=[(x_val1, y_val1)], 
              eval_metric="rmse")
    model2 = LGBMRegressor()
    model2.fit(x_train2, y_train2, 
              callbacks=[lgb.early_stopping(stopping_rounds=10, verbose=True)],
              eval_set=[(x_val2, y_val2)], 
              eval_metric="rmse")
    model3 = LGBMRegressor()
    model3.fit(x_train3, y_train3, 
              callbacks=[lgb.early_stopping(stopping_rounds=10, verbose=True)],
              eval_set=[(x_val3, y_val3)], 
              eval_metric="rmse")
    model4 = LGBMRegressor()
    model4.fit(x_train4, y_train4, 
              callbacks=[lgb.early_stopping(stopping_rounds=10, verbose=True)],
              eval_set=[(x_val4, y_val4)], 
              eval_metric="rmse")
    models.append([model1, model1.score(x_test, y_test)])
    models.append([model2, model2.score(x_test, y_test)])
    models.append([model3, model3.score(x_test, y_test)])
    models.append([model4, model4.score(x_test, y_test)])
```

```python
df_test = pd.read_csv("test(1).csv")
df_test["datetime"] = pd.to_datetime(df_test["datetime"])
df_test.head()
```

```python
df_test["year"] = df_test["datetime"].dt.year
df_test["month"] = df_test["datetime"].dt.month
df_test["day"] = df_test["datetime"].dt.day
df_test.head()
```

```python
x_test = df_test.drop("datetime", axis=1)
x_test.columns
```

```python
models = sorted(models, key=lambda x:x[1], reverse=True)
models[0][1]
```

```python
model1 = models[0][0]
model2 = models[1][0]
model3 = models[2][0]
model4 = models[3][0]
model5 = models[4][0]
y_test1 = model1.predict(x_test)
y_test2 = model2.predict(x_test)
y_test3 = model3.predict(x_test)
y_test4 = model4.predict(x_test)
y_test5 = model5.predict(x_test)
```

```python
y_test = []
data = df_test.values
for i in range(len(data)):
    y_test.append([data[i][0], (y_test1[i]+y_test2[i]+y_test3[i]+y_test4[i]+y_test5[i])/5])
df_post = pd.DataFrame(y_test)
df_post.to_csv("post1.csv", index=False)
```
