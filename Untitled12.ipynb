```python
from sklearn.cluster import KMeans
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = Image.open(".jpg")
im
```

```python
arr = np.array(im)
```

```python
arr
```

```python
data = []
for i in range(len(arr)):
    for j in range(len(arr[i])):
        data.append(arr[i][j])
data = np.array(data)
data
```

```python
model = KMeans(n_clusters=5)
model.fit(data)
```

```python
y_pred = model.predict(data)
plt.hist(y_pred)
```

```python
nrr = []
for i in range(len(y_pred)):
    nrr.append(model.cluster_centers_[y_pred[i]]/255)
nrr
```

```python
arr.shape
```

```python
nrr = np.array(nrr)
nrr = nrr.reshape(191, 264, 3)
```

```python
plt.imshow(nrr)
```

```python
model.cluster_centers_
```

```python
nrr2 = []
for i in range(len(y_pred)):
    if y_pred[i] == 0:
        nrr2.append([255, 255, 255])
    else:
        nrr2.append([0, 0, 0])
nrr2 = np.array(nrr2)
```

```python
nrr2 = nrr2.reshape(191, 264, 3)
```

```python
plt.imshow(nrr2)
```

```python
src = np.zeros((193, 266))
```

```python
for i in range(len(nrr2)):
    for j in range(len(nrr2[i])):
        src[i+1][j+1] = nrr2[i][j][0]
```

```python
dst = []
for i in range(1, len(src)-1):
    tmp = []
    med = []
    for j in range(1, len(src[i])-1):
        med.append([src[i-1][j-1], src[i-1][j], src[i-1][j+1],
                    src[i][j-1], src[i][j], src[i][j+1],
                    src[i+1][j-1], src[i+1][j], src[i+1][j+1]])
        med = np.array(med)
        tmp.append(np.median(med))
        med = []
    dst.append(tmp)
```

```python
plt.imshow(dst, cmap="gray")
```
