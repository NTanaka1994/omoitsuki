```python
from sklearn.cluster import KMeans
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = Image.open(".jpg")
im
```

```python
arr = np.array(im)
```

```python
arr
```

```python
data = []
for i in range(len(arr)):
    for j in range(len(arr[i])):
        data.append(arr[i][j])
data = np.array(data)
data
```

```python
model = KMeans(n_clusters=5)
model.fit(data)
```

```python
y_pred = model.predict(data)
plt.hist(y_pred)
```

```python
nrr = []
for i in range(len(y_pred)):
    nrr.append(model.cluster_centers_[y_pred[i]]/255)
nrr
```

```python
arr.shape
```

```python
nrr = np.array(nrr)
nrr = nrr.reshape(191, 264, 3)
```

```python
plt.imshow(nrr)
```
