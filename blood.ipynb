# 機械学習の準備

```python
from lightgbm import LGBMRegressor as LGBMR
```

```python
x_train = train2.drop("bg+1:00", axis=1)
y_train = train2["bg+1:00"]
x_test = test
```

## 訓練データをダミー変数化

```python
x_train = pd.get_dummies(x_train)
x_train
```

## テストデータのダミー変数化

```python
x_test = pd.get_dummies(x_test)
x_test
```

## テストデータのカラムを合わせる

```python
columns = []
for col in x_train.columns:
    if col in x_test.columns:
        columns.append(col)
```

```python
x_train = x_train[columns]
```

```python
for col in x_test.columns:
    if col not in x_train.columns:
        x_test = x_test.drop(col, axis=1)
```

```python
x_train = x_train.sort_index(axis=1)
x_test = x_test.sort_index(axis=1)
```

```python
if False in (x_train.columns == x_test.columns):
    print("False")
```

## カラムを含めて学習できるようにする

```python
import re
x_train = x_train.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
x_test = x_test.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
model = LGBMR()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
import lightgbm as lgb
for i in range(200):
    params = {'objective': 'regression',  # 最小化させるべき損失関数(2クラス分類の場合'binary'、多クラス分類の場合'multiclass')
            'random_state': i,  # 乱数シード
            'boosting_type': 'gbdt',  # boosting_type
            'n_estimators': 100  # 最大学習サイクル数。early_stopping使用時は大きな値を入力
            }
    x_tr, x_va, y_tr, y_va = train_test_split(x_train, y_train, random_state=i, test_size=0.1)
    model = LGBMR(**params)
    model.fit(x_tr, y_tr,
              eval_metric='rmse',
              eval_set=[(x_va, y_va)], 
              callbacks=[lgb.early_stopping(stopping_rounds=10, 
                                            verbose=True), # early_stopping用コールバック関数
                         lgb.log_evaluation(-1)])# コマンドライン出力用コールバック関数
    models.append([model, model.score(x_va, y_va)])
    print(i+1, model.score(x_va, y_va))
models = sorted(models, key=lambda x:x[1], reverse=True)
```
