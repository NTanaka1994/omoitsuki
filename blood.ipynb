# 機械学習の準備
欠損値が多すぎるためGBDTを使う

```python
from lightgbm import LGBMRegressor as LGBMR
```

```python
x_train = train2.drop("bg+1:00", axis=1)
y_train = train2["bg+1:00"]
x_test = test
```

## 訓練データをダミー変数化

```python
x_train = pd.get_dummies(x_train)
x_train
```

## テストデータのダミー変数化

```python
x_test = pd.get_dummies(x_test)
x_test
```

## テストデータのカラムを合わせる

```python
columns = []
for col in x_train.columns:
    if col in x_test.columns:
        columns.append(col)
```

```python
x_train = x_train[columns]
```

```python
for col in x_test.columns:
    if col not in x_train.columns:
        x_test = x_test.drop(col, axis=1)
```

```python
x_train = x_train.sort_index(axis=1)
x_test = x_test.sort_index(axis=1)
```

```python
if False in (x_train.columns == x_test.columns):
    print("False")
```

## カラムを含めて学習できるようにする

```python
import re
x_train = x_train.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
x_test = x_test.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
model = LGBMR()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

# 予測結果をデータフレーム化して提出用データにする

```python
pred = pd.DataFrame(y_pred)
pred.index = x_test.index
pred.columns = ["bg+1:00"]
pred.sort_index().to_csv("submit_1.csv")
```

```python
pred
```

精度はRMSEで2.5474で10月2日時点で101人中30位


```python
from sklearn.ensemble import GradientBoostingRegressor as GBR
from xgboost import XGBRegressor
```

```python
model1 = LGBMR()
model1.fit(x_train, y_train)
model2 = XGBRegressor()
model2.fit(x_train, y_train)
model3 = GBR()
model3.fit(x_train, y_train)
```

```python
import numpy as np
y_pred = []
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
y_pred3 = model3.predict(x_test)
for i in range(len(y_pred1)):
    y_pred.append(np.mean([y_pred1[i], y_pred2[i], y_pred3[i]]))
pred = pd.DataFrame(y_pred)
pred.index = x_test.index
pred.columns = ["bg+1:00"]
pred.sort_index().to_csv("submit_2.csv")
pred
```

```python
from sklearn.model_selection import train_test_split
models = []
for i in range(50):
    x_tr, x_va, y_tr, y_va = train_test_split(x_train, y_train, random_state=i, test_size=0.1)
    model = LGBMR(verbose=-1)
    model.fit(x_tr, y_tr)
    models.append([model, model.score(x_va, y_va)])
    print(i+1)
```

```python
models = sorted(models, key=lambda x:x[1], reverse=True)
models
```

```python
y_pred = []
y_pred1 = models[0][0].predict(x_test)
y_pred2 = models[1][0].predict(x_test)
y_pred3 = models[2][0].predict(x_test)
for i in range(len(y_pred1)):
    y_pred.append(np.mean([y_pred1[i], y_pred2[i], y_pred3[i]]))
pred = pd.DataFrame(y_pred)
pred.index = x_test.index
pred.columns = ["bg+1:00"]
pred.sort_index().to_csv("submit_3.csv")
pred
```
